```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img("/content/cats.jpg", target_size = (200, 200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    fill_mode = "reflect",
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
type(img)
```

```python
input_batch = img.reshape(1,200,200,3) # here 1 is number of images and remaining is class and shape of the image
```

```python
type(input_batch)
```

```python
i = 0
for output in datagen.flow(input_batch,batch_size = 1, save_to_dir = "/content/aug" ):
  i = i+1
  if i==10:
    break
```
